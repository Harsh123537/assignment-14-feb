Multithreading refers to concurrently executing multiple threads by rapidly switching the control of the CPU between threads.
Multithreading in Python enables CPUs to run different parts(threads) of a process concurrently to maximize CPU utilization.
threading is used to handle threads in python.


threading module is used to implement the multithreading.

1) activeCount()
   activeCount module is used to count the currently active or running threads.
2) currentThread()
   current_thread() is an inbuilt method of the threading module in Python. It is used to return the current Thread      object, which corresponds to the caller's thread of control.
3) enumerate()
    Enumerate() Returns a list of all thread objects that are currently active. 



run() − The run() method is the entry point for a thread.

start() − The start() method starts a thread by calling the run method.

join() − The join() waits for threads to terminate.

isAlive() − The isAlive() method checks whether a thread is still executing.

```python
import threading
def thread1():
    print("list of square")
def thread2():
    print("list of cubes")
th1=threading.Thread(target=thread1,args=())
th2=threading.Thread(target=thread2,args=())
th1.start()
th2.start()
```
